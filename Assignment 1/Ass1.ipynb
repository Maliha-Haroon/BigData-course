# **Distance Conversion**

```python
km = float(input("Enter distance between two cities in kilometers: "))

meters = km * 1000
feet = km * 3280.84
inches = km * 39370.1
centimeters = km * 100000

print("Distance in meters:", meters)
print("Distance in feet:", feet)
print("Distance in inches:", inches)
print("Distance in centimeters:", centimeters)
```

# **Area & Perimeter of Rectangle and Circle**

```python
import math

length = float(input("Enter length of rectangle: "))
breadth = float(input("Enter breadth of rectangle: "))
radius = float(input("Enter radius of circle: "))
```

```python
rect_area = length * breadth
rect_perimeter = 2 * (length + breadth)
```

```python
circle_area = math.pi * radius * radius
circle_circumference = 2 * math.pi * radius
```

```python
print("Rectangle Area:", rect_area)
print("Rectangle Perimeter:", rect_perimeter)
print("Circle Area:", circle_area)
print("Circle Circumference:", circle_circumference)
```

# **Cost Price of One Item**

```python
total_selling_price = float(input("Enter total selling price of 5 items: "))
total_profit = float(input("Enter total profit earned: "))

total_cost_price = total_selling_price - total_profit
cost_price_per_item = total_cost_price / 5

print("Cost Price of One Item:", cost_price_per_item)
```
