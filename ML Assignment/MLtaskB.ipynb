## Task 1a: Decision Tree (Original Parameters)
- Split: 70% train, 30% test
- Criterion: entropy
- Max Depth: 3


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.read_csv("golf_dataset.csv")
```

```python
le = LabelEncoder()
data['Outlook'] = le.fit_transform(data['Outlook'])
data['Wind'] = le.fit_transform(data['Wind'])
data['Play'] = le.fit_transform(data['Play'])
```

```python
X = data.drop('Play', axis=1)
y = data['Play']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
dt_model = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=42)
dt_model.fit(X_train, y_train)
```

```python
y_pred = dt_model.predict(X_test)
print("Decision Tree Accuracy (Task 1):", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, zero_division=0))
```
